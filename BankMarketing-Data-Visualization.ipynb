# **UNIVARIATE ANALYSIS**

Q1. Highest age to have Loan? Is There any downfall after that If Yes, Give Reason.

```python
plt.figure(figsize=(10,10))
sns.countplot(y=df['age'], color='#C9A0DC')
plt.yticks(rotation=20)
plt.title("Age and Indept Loan in account.", fontsize=20)
plt.xlabel("Loan in Thousand", fontsize=20, alpha=0.7)
plt.ylabel("Age", fontsize=20, alpha=0.7)
plt.tight_layout()
plt.show()
# By looking at below graph we can conclude that most number of loan taken in between age of 30-35.
# There comes up a sudden downfall After age 60 which can be the reason of retirement
```

Q2. Which job Have Most Depth in Account? Also check the less Depth in which Job type.

```python
plt.figure(figsize=(12,10))
plt.title('Depth on the basis of Job', fontsize=25)
sns.countplot(data=df,y='job',order=df['job'].value_counts().index[0:15],palette='Accent')
plt.xlabel('Earning', fontsize=20, alpha=0.7)
plt.ylabel('Job', fontsize=20, alpha=0.7)
plt.show()
# Management is at Highest rank for earning in 2020.
# Unknown stands for Lowest earning.
```

# BIVARIATE ANALYSIS

Q3. What are the job with most number of Deposit in 2020? Also find the least one from Top.

```python
deposit = df[df['deposit']=='yes']
plt.figure(figsize=(15,8))
plt.title('No. of Deposit in 2020', fontsize=25)
sns.countplot(y='job',data=deposit,order=deposit['job'].value_counts().index[0:],palette='coolwarm')
plt.xlabel('No. of people who already Deposited', fontsize=20, alpha=0.7)
plt.ylabel('Job', fontsize=20, alpha=0.7)
plt.show()
# The job with most number of deposite is management who has made more than 1300 people in Deposite.
# The job with least number of deposite from Top 10 is Unknown.
```

### Q4. What is the Duration of the Loan?

```python
sns.boxplot(x='loan',y='duration',data=df)
plt.show()
# Higest Duration of the loan is 3300.
```

### Q5. Show Stripplot Visualization for job of the employee and their Duration.

```python
sns.stripplot(x='job',y='duration',data=df,palette='Accent',jitter=0.2)
plt.xticks(rotation=90)
plt.show()
```
